#### QUESTION 2: Read and parse data from the```"ride.csv"```. Complete the given code (8 marks)

````python
import csv

def load_ride_data(filename):
    data = []
    with open(filename, 'r') as file:
        reader = csv.DictReader(file) 
        #read and parse data iteratively  (5 marks)
        for row in reader:
            #convert numeric values to correct data type 
            #to access dict values = row['key']
            row['distance'] = float(row['distance'])
            row['duration'] = int(row['duration'])
            row['fare'] = float(row['fare'])
            data.append(row)
    return data

# Write code(s) to load data from actual CSV file (2 marks)
ride_data = load_ride_data('ride.csv')

# Write code(s) to preview data (1 mark)
print(ride_data[0:2])
````

#### QUESTION 3: Define a function to compute average fare by city and call the function by passing the data (10 marks)

````python
# Write your function definition here to compute average fare by city (8 marks)
def average_fare_by_city(data):
    city_total_fare = {} #store total fare for each city
    city_ride_count = {} #store number of ride for each city

    for ride in data:
        city = ride['city'] #access dict to get city name
        fare = ride['fare'] #access dict to get fare amount

        #if city still dont add yet, start from 0
        if city not in city_total_fare:
            city_total_fare[city] = 0
            city_ride_count[city] = 0

        city_total_fare[city] += fare
        city_ride_count[city] += 1

    average_fare = {}
    for city in city_total_fare:
        average_fare[city] = round(city_total_fare[city] / city_ride_count[city],2)

    for city in city_total_fare:
        print('Total rider for ' + city + ': ' + str(city_ride_count[city]))

    return average_fare 

# Write the code(s) to call your function and pass the data (2 marks)
avrg_fare = average_fare_by_city(ride_data)
print()
for city, fare in avrg_fare.items():
    print(city + ': RM ' + str(round(fare,2)))

````

#### QUESTION 4: Define a function to summarize driver information, call the function by passing data and display the summary (10 marks)
#### ```driver_name, num_of_rides, total_distance, total_fare```

````python
# Write your function definition here (7 marks)
def summarize_driver(data):
    summary = {} #store each driver summarize info
    
    for ride in data:
        driver = ride['driver_name']
        distance = ride['distance']
        fare = ride['fare']
        
        #if driver not added yet, start with 0
        if driver not in summary:
            summary[driver] = {
                'num_of_rides': 0, 'total_distance': 0, 'total_fare': 0
            }
    
        #update the driver information
        summary[driver]['num_of_rides'] += 1
        summary[driver]['total_distance'] += distance
        summary[driver]['total_fare'] += fare

    return summary

# Write the code(s) to call your function and pass the data (2 marks)
summarize_information = summarize_driver(ride_data)

# Write the code(s) to display driver's summary (1 mark)
for driver, info in summarize_information.items():
    print("Driver's name: " + driver)
    print("Number of rides: " + str(info['num_of_rides']))
    print('Total distance: ' + str(info['total_distance']))
    print('Total fare: ' + str(info['total_fare']) + '\n')
````

#### QUESTION 5: Implement a function ```analyze_peak_slice(data, start_idx, end_idx)``` that: (10 marks)

    1.Accepts the ride data list and two indices: start_idx and end_idx.
    2.Extracts a slice of rides from start_idx = 2 to end_idx = 6 (inclusive).
    3.Calculates and returns:
        i)  Total number of rides in that slice
        ii) Average fare
        iii)List of unique drivers involved
        
    Sample output:
 
   ```Total_rides: 5```<br>
   ```Average_fare: 18.45```<br>
   ```Drivers: Ravi, Ali, Mei``` 

````python
# Write your function definition here that accepts ride data, start index and end index, slice the list, 
# calculate and returns the values in (3)   (8 marks)
def analyze_peak_slice(data, start_idx, end_idx):
    #slice data from start to end index(inclusive)
    sliced_data = data[start_idx:end_idx + 1]

    #count number of rides in the slice
    num_rides = len(sliced_data)

    #calculate total fare
    total_fare = sum(ride['fare'] for ride in sliced_data)

    #calculate average fare
    avg_fare = round((total_fare / num_rides), 2)

    #get unique drivers using a set so that no preparation
    driver = set(ride['driver_name'] for ride in sliced_data)

    return num_rides, avg_fare, list(driver)

# Write the code(s) to call your function and pass the ride data, start index and end index (2 marks)
rides, average, drivers = analyze_peak_slice(ride_data,2,6)

print('Total_rides: ', rides)
print(f'Average_fare : {average:.2f}') 
print('Drivers: ', ', '.join(drivers))
````

#### QUESTION 6: Use lambda function to find top 3 drivers by total fare (7 marks)

````python
# Write your function definition here for top 3 drivers by total fare (5 marks)
def top_3_drivers_by_totfare(summary):
    #convert dict to list(driver,info) and sort by tot fare
    sort_driver = sorted(
        summary.items(), #give (driver,info)
        key = lambda item: item[1]['total_fare'], #sort by total_fare
        reverse = True #highest first
    )
    return sort_driver[0:3]

# Write the code(s) to call your function and pass the driver's summary data (2 marks)
summarize_information = summarize_driver(ride_data)
top_drivers = top_3_drivers_by_totfare(summarize_information)

print("Top 3 drivers by total fare:")
print()
for count, (driver, info) in enumerate(top_drivers, 1):
    print('Top ' + str(count) + ': ' + driver)
    print(f'Total fare: RM{info['total_fare']:.2f}')
    print()
````

#### QUESTION 7: Write a function to calculate average duration by cities and use lambda function to sort cities by average duration (10 marks)

````python
# Write your function definition here for average duration and sort cities (8 marks)
def average_duration_by_city(data):
    city_duration_tot = {} #duration per city
    city_ride_count = {} #number of rides per city
    
    for ride in data:
        city = ride['city']
        duration = ride['duration']

        if city not in city_duration_tot:
            city_duration_tot[city] = 0
            city_ride_count[city] = 0
    
        city_duration_tot[city] += duration
        city_ride_count[city] += 1

    average_duration = {}
    for city in city_duration_tot:
        average_duration[city] = city_duration_tot[city]/city_ride_count[city]

    return average_duration

# Write the code(s) to call your function and pass the data (2 marks)
avg_duration = average_duration_by_city(ride_data)
sorted_cities = sorted(avg_duration.items(), key=lambda item: item[1], reverse=True)
print("Cities sorted by average duration:")
for city, duration in sorted_cities:
    print(f"{city}: {duration:.2f} minutes")
````

#### QUESTION 8: Define a function to write driver summary to a text file ```driver_summary.txt``` with the following information:
#### ```"Driver", "Total Rides", "Total Distance", "Total Fare"```

````python
# Write your function definition here for writing driver summary into a text file (8 marks)
def summary_to_file(summary, filename="driver_summary.txt"):
    with open(filename, 'w') as file:
        file.write('Driver\tTotal Rides\tTotal Distance\tTotal Fare\n')

        for driver, info in summary.items():
            line = driver + '\t' + str(info['num_of_rides']) + '\t\t' + str(round(info['total_distance'],2)) + '\t\t' + str(round(info['total_fare'],2)) + '\n'
            file.write(line)

# Write the code(s) to call your function and pass the driver summary data (2 marks)
summarize_information = summarize_driver(ride_data)
summary_to_file(summarize_information)
````
